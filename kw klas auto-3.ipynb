```python
#웹 접속 후 로그인
driver = webdriver.Chrome()
driver.get('https://klas.kw.ac.kr')
time.sleep(3)

ele = driver.find_element(By.CSS_SELECTOR, '#loginId')
ele.send_keys(klas_id +'\t' + klas_pwd +'\n')

time.sleep(1)

#데이터 저장 리스트
data_num = [] #수강 현황
data_title = [] #강좌 제목
data_miss = [] #미수강 강좌 정보 획득


try: 
    for lesson_num in range(1,15):
        #강좌 선택
        sel = f'#appModule > div > div:nth-child(1) > div:nth-child(2) > ul > li:nth-child({lesson_num}) > div.left > span'
        ele = driver.find_element(By.CSS_SELECTOR, sel)
        if ele.text == "서울권역 e-러닝":
            data_num.append('e-러닝')
            data_title.append((driver.find_element(By.XPATH, f'//*[@id="appModule"]/div/div[1]/div[2]/ul/li[{lesson_num}]/div[1]').text.strip("\n서울권역 e-러닝")))            
            lesson_num += 1
        else:
            ele.click()
            time.sleep(1)
            #강좌수 데이터 추출
            data_num.append((driver.find_element(By.XPATH, '//*[@id="appModule"]/div[1]/div[2]/div/div[2]/div/div[1]/ul/li[1]/a/span').text))
            data_title.append((driver.find_element(By.XPATH, '//*[@id="appHeaderSubj"]/div/div/div[1]/span[1]').text))

            #홈으로 가기
            sel_home = 'body > header > div.navbar.navbar-dark.bg-top > div > div.col-md-6.toplogo > a > img'
            ele_home = driver.find_element(By.CSS_SELECTOR, sel_home)
            ele_home.click()
            
            lesson_num += 1
            time.sleep(1)
        
except:
    pass

print(f" [{len(data_num)}개의 데이터 수집완료]\n")



for i in range(len(data_num)):
    print('(', i+1, ') ' + data_num[i] + '\t' + data_title[i])
    
print("\n--------------------------------------------\n")    
for i in range(len(data_num)):
    check = data_num[i]
    if (check == '0/0' or check == '1/1' or check == '2/2' or 
        check == '3/3' or check == '4/4' or check == '5/5' or 
        check == '15/15' or check == '16/16' or check == '17/17' or
        check == '18/18' or check == '19/19' or check == '20/20' or 
        check == '6/6' or check == '7/7' or check == '8/8' or 
        check == '9/9' or check == '10/10' or check == '11/11' or
        check == '12/12' or check == '13/13' or check == '14/14' or 
        check == '21/21' or check == '22/22' or check == '23/23' or 
        check == '24/24' or check == '25/25' or check == '26/26' or 
        check == '27/27' or check == '28/28' or check == '29/29' or 
        check == '30/30' or check == 'e-러닝'):
        pass
    
    else:
        print('(수강이 필요합니다)   ' + data_num[i] + '\t' + data_title[i])
        data_miss.append(i) # 미수강 index 획득
    
    
```
